```python
import pandas as pd
```

```python
#establishing a connection with mysql.
import mysql.connector as mc
```

```python
config = {
    "host" : "localhost",
    "user" : "root",
    "password" : "admin",
    "database" : "employee"
}
```

```python
#creating a connection to the database.
conn = mc.connect(**config)
```

```python
#table name.
tableName = "empDetails"
```

```python
query = f"select *from {tableName}"
```

```python
df = pd.read_sql(query,conn)
```

```python
import matplotlib.pyplot as plt
```

```python
#scatter plot.
x = df.age
y = df.salary
plt.scatter(x,y)
plt.xlabel("age");
plt.ylabel("salary")
plt.show()
```

```python
#bar plot.
x = df.age
y = df.salary
plt.bar(x,y)
plt.xlabel("age");
plt.ylabel("salary")
plt.show()
```

```python
#box plot.
x = df.salary
plt.boxplot(x)
plt.xlabel("salary");
plt.show()
```

```python
#histogram
x = df.salary
plt.hist(x,bins=7)
plt.show()
```

```python
#lineplot
x = df.age
y = df.salary
plt.plot(x,y)
plt.xlabel("age");
plt.ylabel("salary")
plt.show()
```
