```python
#K-Means algorithm
import math
import pandas as pd
```

```python
dp = [(2,10),(2,5),(8,4),(5,8),(7,5),(6,4),(1,2),(4,9)]
```

```python
#consider three centroids a1,b1,c1 initially
c  =[(2,10),(5,8),(1,2)]
```

```python
data = {
    "Distance C1":[None]*len(dp),
    "Distance C2":[None]*len(dp),
    "Distance C3":[None]*len(dp),
    "Cluster":[None]*len(dp),
    "New Cluster":[None]*len(dp)
}
```

```python
df = pd.DataFrame(data,index=['A1','A2','A3','B1','B2','B3','C1','C2'])
```

```python
df
```

```python
while(True):
    for i in range(len(df)):
        row_distance = []
        for j in range(0,len(c)):
            #calculating the eucledian distance.
            x1 = dp[i][0]
            y1 = dp[i][1]
            
            x2 = c[j][0]
            y2 = c[j][1]
            
            ed = ((x2-x1)**2+(y2-y1)**2)**0.5
            ed = round(ed,2)
            df.iloc[i,j] = ed
            row_distance.append(ed)
        min_dist = min(row_distance)
        df.iloc[i,4] = row_distance.index(min_dist)+1
    #check whether new cluster and old cluster values or same or not
    if(df['Cluster'].equals(df['New Cluster'])):  #is equal break the loop.
        print(c)
        print(df)
        break 
    else:
        temp_list = [[0, 0] for _ in range(len(c))]
        for i in range(len(df)):
            cluster_val = df.iloc[i,4] # new cluster value
            x = dp[i][0]
            y = dp[i][1]
            temp_list[cluster_val - 1][0] += x
            temp_list[cluster_val - 1][1] += y
        for i in range(len(temp_list)):
            frequency = df['New Cluster'].value_counts().get(i+1, 0)
            if(frequency>0):
                temp_list[i][0] = temp_list[i][0]/frequency
                temp_list[i][1] = temp_list[i][1]/frequency
        c = temp_list
        #new cluster becomes old cluster, loop repeats again.   
        print(df)
        print(c)
        df['Cluster'] = df['New Cluster']
#     break
            
```
